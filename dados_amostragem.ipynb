```python
import pandas as pd
```

```python
data_amazonas = pd.read_csv("votacao_secao_2022_AM.csv", sep=";", encoding="latin1")
data_roraima = pd.read_csv("votacao_secao_2022_RR.csv", sep=";", encoding="latin1")
```

```python
data_geral = pd.read_csv("votacao_secao_2022_BR.csv", sep=";", encoding="latin1")
```

```python
print(f"Dimensão da base de dados de Amazonas: {data_amazonas.shape} \n Dimensão da base de dados de Roraima: {data_roraima.shape}")
```

```python
data_amazonas.head()
```

```python
data_amazonas_new = data_amazonas[['NR_TURNO', 'DS_CARGO', 'NM_VOTAVEL', 'QT_VOTOS']]
```

```python
data_amazonas_new.head()
```

```python
data_roraima_new = data_roraima[['NR_TURNO', 'DS_CARGO', 'NM_VOTAVEL', 'QT_VOTOS']]
```

```python
data_roraima_new.head()
```

```python
data_geral = data_geral[['NR_TURNO', 'SG_UF' ,'DS_CARGO', 'NM_VOTAVEL', 'QT_VOTOS']]
```

```python
data_geral_presidente = data_geral[data_geral['DS_CARGO'] == 'PRESIDENTE']
```

```python
data_amazonas_governador = data_amazonas_new[data_amazonas_new['DS_CARGO'] == 'GOVERNADOR']
data_amazonas_senador = data_amazonas_new[data_amazonas_new['DS_CARGO'] == 'SENADOR']
data_roraima_senador = data_roraima_new[data_roraima_new['DS_CARGO'] == 'SENADOR']
data_roraima_governador = data_roraima_new[data_roraima_new['DS_CARGO'] == 'GOVERNADOR']
data_presidente = data_geral_presidente[(data_geral_presidente['SG_UF'] == 'AM') | (data_geral_presidente['SG_UF'] == 'RR')].reset_index()

data_amazonas_governador['Estado'] = 'AM'
data_roraima_governador['Estado'] = 'RR'
data_amazonas_senador['Estado'] = 'AM'
data_roraima_senador['Estado'] = 'RR'



data_governador = pd.concat([data_amazonas_governador, data_roraima_governador], ignore_index=True)

data_senador = pd.concat([data_amazonas_senador, data_roraima_senador], ignore_index=True)
```

```python
data_presidente["Estado"] = data_presidente['SG_UF']
data_presidente = data_presidente.drop(columns = ['SG_UF', 'index'])
```

```python
data_governador_expanded = data_governador.loc[data_governador.index.repeat(data_governador['QT_VOTOS'])].reset_index(drop=True)
data_governador_expanded = data_governador_expanded.drop(columns=['QT_VOTOS'])

data_senador_expanded = data_senador.loc[data_senador.index.repeat(data_senador['QT_VOTOS'])].reset_index(drop=True)
data_senador_expanded = data_senador_expanded.drop(columns=['QT_VOTOS'])

data_presidente_expanded = data_presidente.loc[data_presidente.index.repeat(data_presidente['QT_VOTOS'])].reset_index(drop=True)
data_presidente_expanded = data_presidente_expanded.drop(columns=['QT_VOTOS'])
```
